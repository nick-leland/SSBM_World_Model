```python
from huggingface_hub import HfApi, hf_hub_download

repo_id = "erickfm/slippi-public-dataset-v3.7"
api = HfApi()

# List files in the dataset repo and keep only .slp replays
slp_files = [
    filename
    for filename in api.list_repo_files(repo_id=repo_id, repo_type="dataset")
    if filename.endswith(".slp")
]

# Download only the first 5 .slp files
first5_paths = [
    hf_hub_download(repo_id=repo_id, repo_type="dataset", filename=filename)
    for filename in slp_files[:5]
]

first5_paths
```
